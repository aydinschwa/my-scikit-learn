```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from linear_regression import GDRegression
```

# Animating Gradient Descent For Linear Regression

```python
n = 100
X = np.arange(100)
y = X + 50 + np.random.normal(loc=10, scale=15, size=n)
reg = GDRegression(animate=True)
X = np.array(X).reshape(len(X), -1)
reg.fit(X, y)

fig, ax = plt.subplots()
plt.scatter(X, y)
line, = ax.plot([], [], color="red", lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    w = reg.w_history[i]
    b = reg.b_history[i]
    y_line = w * X[:, 0] + b 
    line.set_data(X[:, 0], y_line) 
    return line,

ani = FuncAnimation(fig, animate, frames=len(reg.w_history), init_func=init, blit=True)
HTML(ani.to_jshtml()) # Use this line for displaying the animation in a Jupyter Notebook
# ani.save('animation.mp4').
```
